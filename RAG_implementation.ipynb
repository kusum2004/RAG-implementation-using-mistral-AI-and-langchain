```python
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader
from llama_index.llms.huggingface import HuggingFaceLLM
```

```python
!mkdir data
```

```python
#load documents
documents = SimpleDirectoryReader("/content/data").load_data()
```

```python
documents
```

```python
#setup prompts - specific to StableLM
from llama_index.core import PromptTemplate

system_prompt = """<|SYSTEM|>#You are a Q&A assistane. Your goal is to answer question as
accurately as possible based on the instructions and context provided.
"""
#This will wrap the default propmts that are internal to llama-index
query_wrapper_prompt = PromptTemplate("<|USER|>{query_str}\n<|ASSISTANT|>")
```

```python
from huggingface_hub import login

# paste your token here (or use input() for safety)
login("your_huggingface id")
```

```python
import torch

llm= HuggingFaceLLM(
    context_window=4096,
    max_new_tokens=256,
    generate_kwargs={"temperature":0.0, "do_sample":False},
    system_prompt=system_prompt,
    query_wrapper_prompt=query_wrapper_prompt,
    tokenizer_name="mistralai/Mistral-7B-Instruct-v0.1",
    model_name="mistralai/Mistral-7B-Instruct-v0.1",
    stopping_ids=[50278, 50279, 50277, 1, 0],
    tokenizer_kwargs={"max_length": 4096},
    model_kwargs={"torch_dtype": torch.float16}
)
```

```python
%pip install llama-index-embeddings-huggingface
%pip install llama-index-embeddings-instructor
```

```python
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
embed_model = HuggingFaceEmbedding(model_name="sentence-transformers/all-mpnet-base-v2")
```

```python
from llama_index.core import VectorStoreIndex, Settings

Settings.chunk_size = 1024
Settings.llm = llm
Settings.embed_model = embed_model
```

```python
index = VectorStoreIndex.from_documents(documents)
```

```python
query_engine = index.as_query_engine()
```

```python
query_engine.query("what skills kusuma have?")
```
